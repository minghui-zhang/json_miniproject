## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

```python
# 2. Find the top 10 major project themes (using column 'mjtheme_namecode')

# the mjtheme_namecode column contains lists of dictionaries that needs to be normalized
json_df.loc[0, 'mjtheme_namecode']

# to normalize the data, need to convert data frame to list of dictionaries
list_of_dicts = json_df.to_dict('records')

# normalize to the project themes
normalized = json_normalize(list_of_dicts, 'mjtheme_namecode')

# group by the name, count rows, sort, and take the top 10 rows
top_projects = normalized.groupby('name').size().sort_values(ascending = False).head(10)

# because some of the project names are missing, group by project code as well
top_projects_code = normalized.groupby('code').size().sort_values(ascending = False).head(10)

print(top_projects)
print(top_projects_code)

top_projects.plot(kind = 'bar')
plt.ylabel('Number of Projects')
plt.title('Grouping by name of project')
plt.show()

top_projects_code.plot(kind = 'bar')
plt.ylabel('Number of Projects')
plt.title('Grouping by project code')
plt.show()
```

```python
# 3. Create a dataframe with the missing names filled in.

# create a mapping of the project code to its name
code_lookup_dict = normalized.drop_duplicates().sort_values('code').replace('', np.nan).dropna().to_dict(orient = 'records')

# create a dictionary of dictionaries in which the code value serves as the key in outer dictionary
code_lookup_nesteddict = {}
for item in code_lookup_dict:
   code = item['code']
   code_lookup_nesteddict[code] = item

# loop through each row in json_df, pick out the project code column
for index, row in json_df.iterrows():
    code_row = row['mjtheme_namecode']
    
    # loop through the list of dictionaries in the code entry
    for dict_index, dictionary in enumerate(code_row):
        
        # if the dictionary contains empty 'name' key, replace the dictionary with appropriate entry in code_lookup
        if dictionary['name'] == '':
            replace_code = dictionary['code']
            replace_dict = code_lookup_nesteddict[replace_code]
            
            # do the replacement
            json_df.iloc[index]['mjtheme_namecode'][dict_index] = replace_dict

# check that all the empty strings are replaced by plotting number of ALL projects (not just top 10)
# to normalize the data, need to convert data frame to list of dictionaries
list_of_dicts = json_df.to_dict('records')

# normalize to the project themes
normalized = json_normalize(list_of_dicts, 'mjtheme_namecode')

# group by the name, count rows, sort
top_projects = normalized.groupby('name').size().sort_values(ascending = False)

top_projects.plot(kind = 'bar') # this plot matches the second plot from #2 (grouping by project code)
plt.ylabel('Number of Projects')
plt.title('Grouping by name of project')
plt.show()
```
